Plot training and validation losses from our model:

```python
from configs.cfg_importer import import_cfg
import matplotlib.pyplot as plt
import json

date = "08_02_2026_20_22_42" # Enter training date to plot correct metrics
model = "lmc"
num_epochs = 104 # Enter here if early stop
with open(f'../logs/model-metrics/model_metrics_{model}_{date}.json', 'r') as f:
    metrics = json.load(f)

train_losses = metrics['training_loss']
val_losses = metrics['validation_loss']

cfg = import_cfg()

plt.figure(figsize=(10, 6))
plt.plot(range(1, num_epochs + 1), train_losses, label="Training Loss")
plt.plot(range(1, num_epochs + 1), val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```
